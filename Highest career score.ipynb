```python
import pandas as pd
```

```python

# Data Extracted from
df=pd.read_csv('Book1.csv')
df.info()
```

```python
df.isna().any()
```

```python
df.rename(columns={'Runs':'Total_Runs_in_Career',"Mat":"Matches","NO":"Not_outs","HS":"Highest_Score","Ave":"Average_Score","BF":"Balls_Faced","SR":"Strike_Rate"},inplace=True)

df.head(15)

```

```python
# df["Span"].str.split("-")


df["Rookie_year"]=df["Span"].str.split("-").str[0]
df['Retirement_year']=df["Span"].str.split("-").str[1]

df
```

```python
df['Rookie_year']=df['Rookie_year'].astype('int32')

df['Retirement_year']=df['Retirement_year'].astype('int32')

df
```

```python
# df['Player'].str.split('(').head()
df['Player_Name']=df['Player'].str.split('(').str[0]
df['Country']=df['Player'].str.split('(').str[1]

df
```

```python
#run only once
df.drop('Span',axis=1,inplace=True)
```

```python

df['Country']=df['Country'].str.strip(')')
df.set_index('Player_Name')

```

```python

#run only once
df.drop(columns='Player',axis=1,inplace=True)
```

```python
def High_not_out(score_str):
    if(score_str.find('*')>=0):
       return 1
    else:
        return 0
```

```python
df["Highest_Score_Not_out"]=df['Highest_Score'].apply(High_not_out)

df
```

```python
df['Highest_Score']=df['Highest_Score'].str.strip("* ")
df['Highest_Score']=df['Highest_Score'].astype('int32')
```

```python
def Retired_or_not(score_str):
    if(score_str.find('*')>=0):
       return 1
    else:
        return 0
```

```python
df['Matches']=df['Matches'].astype('object')
df.info()

```

```python
df['Retired_or_not']=df['Matches'].apply(Retired_or_not)
```

```python
df['Matches']=df['Matches'].str.strip('*')

df['Matches']=df['Matches'].astype('int32')
```

```python
df.info()
```

```python
df = df.reindex(columns=['Player_Name','Country','Rookie_year','Retirement_year','Matches','Inns','Not_outs','Total_Runs_in_Career','Highest_Score','Average_Score','Balls_Faced','Strike_Rate','100','50','0','4s','6s','Highest_Score_Not_out','Retired_or_not'])
df
```

```python


df.sort_values('Total_Runs_in_Career',ascending=False)
```

```python
df['Total_Runs_in_Career'].max()

df.to_csv("Result.csv")
```
